```python
import time as t
```

```python
x0,x1,x2=1,2,3
b,w1,w2= 0.1,0.2,0.3
```

```python
x=[x0,x1,x2]
w=[b, w1,w2]
```

```python
%%time
def forloop(x,w):
    z=0
    for n,i in enumerate(x):
        z+=i*w[n]

    print(z)
```

```python
%%time
def listcomp(x,w):
    z=sum(i*j for i,j in zip(x,w))
    print(z)
```
